```python
import os
from PIL import Image

def find_best_size(dataset_path, max_size):
    """
    This function finds the best size for scaling images in a dataset to minimize the output dataset size.

    Parameters:
    dataset_path (str): The path to the dataset folder.
    max_size (int): The maximum size of the scaled images.

    Returns:
    best_size (tuple): The best size for scaling the images.
    """
    sizes = range(1, max_size+1)
    total_sizes = []
    for size in sizes:
        total_size = 0
        for filename in os.listdir(dataset_path):
            if filename.endswith('.jpg') or filename.endswith('.png'):
                img_path = os.path.join(dataset_path, filename)
                img = Image.open(img_path)
                img_resized = img.resize((size, size))
                total_size += img_resized.size[0] * img_resized.size[1]
        total_sizes.append(total_size)
    best_size = sizes[total_sizes.index(min(total_sizes))]
    return (best_size, best_size)

path = '/content/drive/My Drive/Ai TP/dataset_sport/'


find_best_size(path, 500)
```

```python

import os
import torchvision
import torchvision.transforms as transforms

def scale_images(dataset_path, output_path, size):
    """
    This function scales all images in a dataset to the same dimension.

    Parameters:
    dataset_path (str): The path to the dataset folder.
    output_path (str): The path to the output folder.
    size (tuple): The desired size of the scaled images.

    Returns:
    None
    """
    transform = transforms.Compose([
        transforms.Resize(size),
        transforms.ToTensor()
    ])

    for filename in os.listdir(dataset_path):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            img_path = os.path.join(dataset_path, filename)
            img = Image.open(img_path)
            img_resized = transform(img)
            output_filename = os.path.join(output_path, filename)
            torchvision.utils.save_image(img_resized, output_filename)


path = '/content/drive/My Drive/Ai TP/dataset_sport/'
path_out = '/content/drive/My Drive/Ai TP/crop_image/'


scale_images(path,path_out,(350,300))
```
