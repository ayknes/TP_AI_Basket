```python
from torchvision import transforms
from PIL import Image
import os

# Input directory containing PNG images
input_dir = '/content/drive/My Drive/Ai TP/dataset_sport/'

# Output directory for saving JPG images
output_dir = '/content/drive/My Drive/Ai TP/crop_image/'
os.makedirs(output_dir, exist_ok=True)

# List all files in the input directory
png_files = [f for f in os.listdir(input_dir) if f.endswith('.png')]

png_to_jpg_transform = transforms.Compose([
    transforms.ToTensor(),  # Convert PIL Image to PyTorch tensor
    transforms.Lambda(lambda x: x[:3, :, :]),  # Remove alpha channel if present
    transforms.ToPILImage(mode='RGB')  # Convert PyTorch tensor to PIL Image
])


# Convert each PNG image to JPG and save in the output directory
for png_file in png_files:
    png_path = os.path.join(input_dir, png_file)
    output_path = os.path.join(output_dir, png_file.replace('.png', '.jpg'))

    # Open the PNG image
    png_image = Image.open(png_path)

    # Convert PNG to JPG
    jpg_image = png_to_jpg_transform(png_image)

    # Save the JPG image
    jpg_image.save(output_path)

    print(f"Converted {png_path} to {output_path}")
```
